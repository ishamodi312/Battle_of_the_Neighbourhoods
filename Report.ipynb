## 1. Introduction

### 1.1 Background

When looking for a place to live, a person's decision of whether or not to select a house is based on its rent, as well as on the availability of various services close to the place. If all the amenities are present in a short enough radius, the person may accept a slightly higher rent. On the other hand, if there are no services nearby, they will have a hard time choosing the house.

A cost-effectiveness matrix will prove very useful in choosing a locality as it will give a services to rent ratio, which can be used to determine the best place to rent an apartment.

### 1.2 Problem

To find the cost effectiveness of rental neighbourhoods in Bangalore on the basis of rent and local amenities.

### 1.3 Interest

This project will be useful to everyone who is looking for the best area in which to look for a home, as it reduces the effort and finds the most convenient area based on nearby services.

## 2. Data Acquisition and Cleaning

### 2.1 Data Sources

#### Rent Data

The average rent of 1-,2-,and 3-BHK apartments is obtained from makaan.com
url: https://www.makaan.com/price-trends/property-rates-for-rent-in-bangalore

The data from the first three pages of the website is used.

#### Amenities Data

The nearby amenities in an area are retrieved using the Foursquare API

### 2.2 Data Cleaning

The average rent of each locality is taken. The rent range is discarded.

All the rows with empty cells have been removed to give a resultant DataFrame of 111 rows and 8 columns.

### 2.3 Feature Selection

The final DataFrame contailns the following columns:

1. locality name
2. average rent of 1BHK apartments
3. average rent of 2BHK apartments
4. average rent of 3BHK apartments
5. and 6. The Latitude and Longitude of area
7. Count of venues selling food
8. Count of entertainment venues

## 3. Methodology

### 3.1 Relation Between rent of different types of apartments and venue count

Scatter plots were created with venue count as independent variable and rent as dependent variable.

The plots show an undetermined relation between rent and food venues, but also show that rent is always positively dependent on entertainment venue count.

### 3.2 The  Cost-Effectiveness Matrix

The cost-effectiveness was calculated for all 3 types of rent according to the following formula:

Service Score = sqrt( food^2  + entertainment ^2 )
Rent Score = rent/10
Cost Effectiveness = Service Score/Rent Score
